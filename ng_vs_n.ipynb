```python
import numpy as np
import matplotlib.pyplot as plt
```

We plot the group index (speed of shape of wave) vs the refractive index for silicon. We use the Sellmeier model, details of which dont concern us.

```python
# Constants
wavelengths = [x+1500 for x in range(100)]
wavelengths = np.multiply(wavelengths,1e-3)

lambda1 = 1.1071 #
ep = 11.6858
A = 0.939816
B = 8.10461e-3
```

```python
n = list()

for i,j in enumerate(wavelengths):
    n.append(np.sqrt(ep + A/wavelengths[i]**2 + B * lambda1**2 / (wavelengths[i]**2 - lambda1**2) ) ) # This equation was given and isn't significant to us
```

```python
# A plot of refractive index against wavelength, for silicon. 
plt.plot(wavelengths,n)
plt.ylabel("Refractive index")
plt.xlabel("Wavelength (Î¼m)")
plt.show()
```

```python
# We approximate the derivative of refrative index against wavelength
dndl = list()

for i in range(len(wavelengths) - 3):
    dndl.append( (n[i+3]-n[i+1]) / (wavelengths[i+3]-wavelengths[i+1]) )
```

We can use the equation between the refractive index $n$ and the group index $n_g$
$$n_g = n - \lambda\frac{dn}{d\lambda}$$
to get the group index.

```python
ng = list()

for i in range(len(wavelengths) - 3):
    ng.append((n[2+i] - wavelengths[i+2] * dndl[i]))
```

```python
plt.plot(wavelengths[3:],ng,label="$n_g$")
plt.plot(wavelengths,n,label="$n$")
plt.ylabel("Refractive index")
plt.xlabel("Wavelength (Î¼m)")
plt.legend()
plt.show()
```

For silicon, as the group index is higher than the refractive index, we see material dispersion.
